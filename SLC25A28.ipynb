```python
#O script tem como principal função criar um ficheiro, chamado "pesquisa.txt", no qual apresenta uma pesquisa no pubmed do texto "SLC25A28, melanoma".
```

```python
from Bio import Entrez
from Bio import Medline

Entrez.email = "beatrizmariarm@gmail.com"
handle = Entrez.egquery(term = "SLC25A28, melanoma")
record = Entrez.read(handle)


for row in record["eGQueryResult"]:
    if row["DbName"]=="pubmed":
        y = row["Count"]
        

handle = Entrez.esearch(db = "pubmed", term = "SLC25A28, melanoma", retmax=y)
record = Entrez.read(handle)
idlist = record["IdList"]


handle = Entrez.efetch(db="pubmed", id=idlist, rettype="medline", retmode="text")
records = list(Medline.parse(handle))
record_results = open('pesquisa.txt', 'w')
for record in records:
    titulo=('Titulo: ', record.get('TI', '?'))
    autores=('Autores: ', record.get('AU', '?'))
    fonte=('Fonte: ', record.get('SO', '?'))
    
    record_results.write(str(titulo))
    record_results.write("\n")
    record_results.write(str(autores))
    record_results.write("\n")
    record_results.write(str(fonte))
    record_results.write("\n")
    record_results.write("\n")
record_results.close()
```

```python
#O ficheiro "sequenceslc.fasta" contém a sequência de aminoácidos da proteína do FDX1. Este ficheiro é necessário para o correr os próximos dois scripts
#O primeiro script dá informações sobre o tamanho da sequência e corre o Blast para realizar uma pesquisa com recurso ao SwissProt por forma a encontrar sequências similares, criando um ficheiro .xml. Seguidamente o programa realiza o parsing com recurso ao ficheiro criado anteriormente, sendo analisados vários parâmetros como a base de dados utilizada, a matriz de substituição e os parâmetros que definem os espaçamentos. São identidicados accesion numbers, e-values e os comprimentos dos alinhamentos que foram devolvidos.
#O procedimento é repetido para o organismo "Homo Sapiens" e os resultados são novamente guardados em ficheiro .xml. É identificado o alinhamento com menor e-value e identificado o registo da Uniprot da sequência e o número de HSP's, analisando posteriormente o mesmo para todos os alinhamentos.
```

```python
from Bio.Blast import NCBIXML 
from Bio.Blast import NCBIWWW 
from Bio import SeqIO

#abre o ficheiro e vê o tamanho da sequência
seqrecord = SeqIO.read(open("sequenceslc.fasta"), format="fasta") 
print (len(seqrecord.seq))
  
#pesquisa sequências similares através do blast, com recurso à swissprot;
#o resultado obtido é guardado num ficheiro xml

result_handle = NCBIWWW.qblast("blastp", "swissprot", seqrecord.format("fasta"))

save_file = open("blast-slc.xml", "w")
save_file.write(result_handle.read()) 
save_file.close() 
result_handle.close()

#parsing
result_handle = open("blast-slc.xml")
record = NCBIXML.read(result_handle)


print ("PARAMETERS:")
print ("Database: " + record.database)
print ("Matrix: " + record.matrix)
print ("Gap penalties: ", record.gap_penalties)

#características dos alinhamentos
nhits = len(record.alignments) 
print ("number hits: ", nhits)

res = []
for alignment in record.alignments:
    evalue = alignment.hsps[0].expect
    accession = alignment.accession
    leng = alignment.hsps[0].align_length
    res.append(accession + " - " + str(evalue) + " length:" + str(leng) )

print("E-values and length of alignments:")
for s in res: print(s)

#blast para Homo Sapiens
result_handle2 = NCBIWWW.qblast("blastp", "swissprot", seqrecord.format("fasta"), entrez_query = "Homo sapiens[organism]" )

save_file = open("blast-slc.xml", "w")
save_file.write(result_handle2.read()) 
save_file.close() 
result_handle2.close()

#análise dos resultados
result_handle2 = open("blast-slc.xml")
record2 = NCBIXML.read(result_handle2)

first_alignment = record2.alignments[0]

print ("Accession: " + first_alignment.accession)
print ("Hit id: " + first_alignment.hit_id)
print ("Definition: " + first_alignment.hit_def)

print ("Number of HSPs: ", len(first_alignment.hsps))

for hsp in first_alignment.hsps:
    print ("E-value: ", hsp.expect)
    print ("Length: ", hsp.align_length)
    print ("Identities: ", hsp.identities)
    print ("Query start: ", hsp.query_start)
    print ("Sbjct start: ", hsp.sbjct_start)
    print (hsp.query[0:90])
    print (hsp.match[0:90])
    print (hsp.sbjct[0:90])
    print ("")
```

```python
#O último script guarda em ficheiro fasta as sequências homólogas obtidas para poderem servir de input em programas como o MEGAX, de construção filogenética.
```

```python
from Bio.Blast import NCBIXML 
from Bio.Blast import NCBIWWW 
from Bio import SeqIO
from Bio import Entrez

seqrecord = SeqIO.read(open("sequenceslc.fasta"), format="fasta")

result_handle = NCBIWWW.qblast("blastp", "swissprot", seqrecord.format("fasta"))

save_file = open("blast-slc.xml", "w")
save_file.write(result_handle.read()) 
save_file.close() 
result_handle.close()

result_handle = open("blast-slc.xml")
record = NCBIXML.read(result_handle)

nhits = len(record.alignments) 
print ("number hits: ", nhits)

Entrez.email = 'beatrizmariarm@gmail.com'

res = []
for alignment in record.alignments:
    evalue = alignment.hsps[0].expect
    accession = alignment.accession
    leng = alignment.hsps[0].align_length
    
    #guardar em ficheiro fasta as sequências homólogas obtidas
    handle = Entrez.efetch(db="protein", id=accession, rettype="fasta", retmode="text")
    seq_record = SeqIO.read(handle, "fasta")
    SeqIO.write(seq_record, accession + '.fasta', "fasta") 
    handle.close()
    
    res.append(accession + " - " + str(evalue) + " length:" + str(leng) )

print("E-values and length of alignments:")
for s in res: print(s)
```
